```python
import os
import pandas as pd
from google.cloud import bigquery

# 1. Peg på din JSON-nøglefil
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/Users/kasperhansen/Desktop/Gdelt/keen-diode-493214-v8-959e0b8e9bdd.json'

# 2. Initialiser klienten
client = bigquery.Client(project='keen-diode-493214-v8')

# Her bruger vi gdelt-bq (officiel) og _PARTITIONDATE for lynhurtig (og gratis) søgning
query = """
SELECT 
    DocumentIdentifier AS Artikel_Link,
    V2Locations AS Lokationer
FROM `gdelt-bq.gdeltv2.gkg_partitioned`
WHERE _PARTITIONDATE >= DATE_SUB(CURRENT_DATE(), INTERVAL 2 DAY)
  AND V2Locations LIKE '%#US#%' 
  AND V2Themes LIKE '%ECON_INFLATION%' -- <--- LIGE HER SKRIVER DU TEMAET!
LIMIT 10
"""

query_job = client.query(query)
df = query_job.to_dataframe(create_bqstorage_client=False)

print(df)
```
